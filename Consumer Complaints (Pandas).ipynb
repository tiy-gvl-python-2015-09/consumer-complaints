```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
%matplotlib inline

complaints = pd.read_csv("consumer-complaints/complaints_dec_2014.csv")
complaints
```

```python
complaints_product = complaints["Product"].value_counts()
```

```python
complaints_product.plot(kind="bar", title="Complaints by Product")
plt.show()
```

```python
complaints_company = complaints["Company"].value_counts()[0:9]
complaints_company.plot(kind="bar", title="Complaints by Company")
plt.show()
```

```python
complaints_response = complaints["Company response"].value_counts()
complaints_response.plot(kind="bar", title="Complaints by Company Response")
plt.show()
```

```python
complaints.insert(8, "Day of Week", pd.to_datetime(complaints["Date received"], format="%m/%d/%Y"))
complaints[["Day of Week"]] = complaints["Day of Week"].dt.weekday
complaints[["Date received"]] = pd.to_datetime(complaints["Date received"], format="%m/%d/%Y")

sum_weekday = complaints["Day of Week"].value_counts()
sum_weekday
```

```python
complaints.index = complaints["Day of Week"]
complaints
```

```python
def get_days_in_month():
    weekdays = {}
    for x in range(0,7):
        weekdays.update({x: len(complaints[["Date received"]].ix[x]["Date received"].value_counts())})
    return weekdays
#complaints_product = complaints["Product"].value_counts()
#len(complaints[["Date received"]].ix[3]["Date received"].value_counts())
num_weekdays = get_days_in_month()
num_weekdays
```

```python
def average_by_day(a, b):
    daily_ave = {}
    for x in range(0, 7):
        daily_ave.update({x: a.get(x) / b.get(x)})
    return daily_ave

print(average_by_day(sum_weekday, num_weekdays))
```

```python
complaints_weekday = average_by_day(sum_weekday, num_weekdays)
plt.bar(range(len(complaints_weekday)), complaints_weekday.values(), align='center')
plt.xticks(range(len(complaints_weekday)), list(["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"]))
plt.ylabel("Average Number of Complaints")
plt.xlabel("Day of Week")
plt.title("Complaints by Day of Week")
plt.show()
```

This is interesting. Complaints came in much more often during the beginning of the week. Sunday received over 500 complaints a day, while Saturday barely received over 100 complaints. 

Christmas fell on a Thursday in 2014, but a cursory look does not reveal any major influence on the data. An interesting follow-up project would be to examine the number of complaints by week, to see how the holidays affect complaints.
